```python
import pandas as pd
import seaborn as sns
```

```python
# load datasets
data = sns.load_dataset("titanic")
data.head()
```

```python
# total number contingency table

pd.crosstab(data["survived"], data["pclass"])
```

###  Joint Probability

```python
joint_probability = pd.crosstab(data["survived"], data["pclass"], normalize="all")

joint_probability
```

### Marginal Probability

```python
marginal_probability = pd.crosstab(data["survived"], data["pclass"], normalize="all", margins= True)

marginal_probability
```

### Conditional Probability

```python
# probability when given the class
conditional_probability = pd.crosstab(data["survived"], data["pclass"], normalize="columns")

conditional_probability
```

```python
# probability when given the survived or not
con_probability = pd.crosstab(data["survived"], data["pclass"], normalize="index")

con_probability
```
